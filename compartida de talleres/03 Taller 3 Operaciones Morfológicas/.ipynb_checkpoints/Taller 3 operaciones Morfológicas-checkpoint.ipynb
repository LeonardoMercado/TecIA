```python
import numpy as np
import cv2 
from matplotlib import pyplot as plt
print('Librerias importadas correctamente')
```

```python
img_1 = cv2.imread('Fig/manoP.png',0)
img_2 = cv2.imread('Fig/manoG.jpg',0)
ret,img_1_bin = cv2.threshold(img_1,127,255,cv2.THRESH_BINARY_INV)
ret,img_2_bin = cv2.threshold(img_2,165,255,cv2.THRESH_BINARY_INV)
print('Las imagenes a trabajar son:')

plt.figure(1)
plt.subplot(2,2,1), plt.imshow(img_1,cmap='gray'),plt.title('Mano 1')
plt.subplot(2,2,2), plt.imshow(img_2,cmap='gray'),plt.title('Mano 2')
plt.subplot(2,2,3), plt.imshow(img_1_bin,cmap='gray'),plt.title('Mano 1 Binaria')
plt.subplot(2,2,4), plt.imshow(img_2_bin,cmap='gray'),plt.title('Mano 2 Binaria')
plt.subplots_adjust(left = 0.125, right = 1.5 ,  bottom = 0.1, top = 1.8, wspace = 0.2, hspace = 0.2 )
plt.show()
```

## Operaciones Básicas

### 1. Dilatación

```python
kernel1 = np.ones((3,3))
dilatation_1 = cv2.dilate(img_1_bin,kernel1,iterations=1)
dilatation_2 = cv2.dilate(img_2_bin,kernel1,iterations=1)
dilatation_10 = cv2.dilate(img_1_bin,kernel1,iterations=10)
dilatation_20 = cv2.dilate(img_2_bin,kernel1,iterations=10)
dilatation_100 = cv2.dilate(img_1_bin,kernel1,iterations=50)
dilatation_200 = cv2.dilate(img_2_bin,kernel1,iterations=100)

print('IMAGENES ORIGINALES')
plt.figure(1)
plt.subplot(1,2,1), plt.imshow(img_1_bin,cmap='gray'), plt.title('Mano 1')
plt.subplot(1,2,2), plt.imshow(img_2_bin,cmap='gray'), plt.title('Mano 2')
plt.subplots_adjust(right = 1.5)
plt.show()

print('SE DILATA CON ELEMENTO ESTRUCTURANTE CUADRADO: ')
plt.figure(2)
plt.subplot(1,2,1), plt.imshow(dilatation_1,cmap='gray'), plt.title('#iteraciones = 1')
plt.subplot(1,2,2), plt.imshow(dilatation_2,cmap='gray'), plt.title('#iteraciones = 1')
plt.subplots_adjust(right = 1.5)
plt.show()


plt.figure(3)
plt.subplot(1,2,1), plt.imshow(dilatation_10,cmap='gray'), plt.title('#iteraciones = 10')
plt.subplot(1,2,2), plt.imshow(dilatation_20,cmap='gray'), plt.title('#iteraciones = 10')
plt.subplots_adjust(right = 1.5)
plt.show()


plt.figure(4)
plt.subplot(1,2,1), plt.imshow(dilatation_100,cmap='gray'), plt.title('#iteraciones = 50')
plt.subplot(1,2,2), plt.imshow(dilatation_200,cmap='gray'), plt.title('#iteraciones = 100')
plt.subplots_adjust(right = 1.5)
plt.show()
```

Notas:  
- Como se observa el efecto de dilatar las imagenes con el elemento estructurante que denominamos para este caso kernel1, hace que la mano "se engrose", o mejor dicho, se extiendan las partes blancas sobre las negras.  
- Para la dilatación con una iteración se observa que para las dos manos se alcanza a cubrir parte de un "hueco" que tenia, y ya en 10 iteraciones estos "huecos" se cubren por completo. 
- Se observa también que a medida que se itera más, la dilatación hace que el contorno se ensanche más y más llegando a perderse por completo el contorno de los dedos. 

```python
kernel_rect = cv2.getStructuringElement(cv2.MORPH_RECT,(3,8),(1,1))
kernel_elip = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(8,3),(1,1))
kernel_cross = cv2.getStructuringElement(cv2.MORPH_CROSS,(5,5),(2,2))
print('Kerner rectangular')
print(kernel_rect)
print('----------')
print('Kerner elíptico')
print(kernel_elip)
print('----------')
print('Kerner Cruz')
print(kernel_cross)
```

En la siguiente etapa se examinará como afectan los distintos elementos estructurantes a la dilatación de las imagenes.  

```python
dilatation_rect_1 = cv2.dilate(img_2_bin,kernel_rect,iterations=1)
dilatation_elip_1 = cv2.dilate(img_2_bin,kernel_elip,iterations=1)
dilatation_cross_1 = cv2.dilate(img_2_bin,kernel_cross,iterations=1)
dilatation_rect_2 = cv2.dilate(img_2_bin,kernel_rect,iterations=10)
dilatation_elip_2 = cv2.dilate(img_2_bin,kernel_elip,iterations=10)
dilatation_cross_2 = cv2.dilate(img_2_bin,kernel_cross,iterations=10)
dilatation_rect_3 = cv2.dilate(img_2_bin,kernel_rect,iterations=100)
dilatation_elip_3 = cv2.dilate(img_2_bin,kernel_elip,iterations=100)
dilatation_cross_3 = cv2.dilate(img_2_bin,kernel_cross,iterations=100)
dilatation_rect_4 = cv2.dilate(img_2_bin,kernel_rect,iterations=250)
dilatation_elip_4 = cv2.dilate(img_2_bin,kernel_elip,iterations=250)
dilatation_cross_4 = cv2.dilate(img_2_bin,kernel_cross,iterations=250)


plt.figure(1)
plt.subplot(1,1,1), plt.imshow(img_2_bin,cmap='gray'),plt.title('imagen original')
plt.show()

plt.figure(2)
print('SE DILATA CON ELEMENTO ESTRUCTURANTE RECTANGULAR')
plt.subplot(1,4,1), plt.imshow(dilatation_rect_1,cmap='gray'), plt.title('# Itereciones= 1');
plt.subplot(1,4,2), plt.imshow(dilatation_rect_2,cmap='gray'), plt.title('# Itereciones= 10');
plt.subplot(1,4,3), plt.imshow(dilatation_rect_3,cmap='gray'), plt.title('# Itereciones= 100');
plt.subplot(1,4,4), plt.imshow(dilatation_rect_4,cmap='gray'), plt.title('# Itereciones= 250');
plt.subplots_adjust(right = 1.5)
plt.show()

plt.figure(3)
print('SE DILATA CON ELEMENTO ESTRUCTURANTE ELÍPTICO')
plt.subplot(1,4,1), plt.imshow(dilatation_elip_1,cmap='gray'), plt.title('# Itereciones= 1');
plt.subplot(1,4,2), plt.imshow(dilatation_elip_2,cmap='gray'), plt.title('# Itereciones= 10');
plt.subplot(1,4,3), plt.imshow(dilatation_elip_3,cmap='gray'), plt.title('# Itereciones= 100');
plt.subplot(1,4,4), plt.imshow(dilatation_elip_4,cmap='gray'), plt.title('# Itereciones= 250')
plt.subplots_adjust(right = 1.5)
plt.show()

plt.figure(4)
print('SE DILATA CON ELEMENTO ESTRUCTURANTE CRUZ')
plt.subplot(1,4,1), plt.imshow(dilatation_cross_1,cmap='gray'), plt.title('# Itereciones= 1');
plt.subplot(1,4,2), plt.imshow(dilatation_cross_2,cmap='gray'), plt.title('# Itereciones= 10');
plt.subplot(1,4,3), plt.imshow(dilatation_cross_3,cmap='gray'), plt.title('# Itereciones= 100')
plt.subplot(1,4,4), plt.imshow(dilatation_cross_4,cmap='gray'), plt.title('# Itereciones= 250');
plt.subplots_adjust(right = 1.5)
plt.show()
```

Nota:  
- Como se observa el elemento estructurante determina en gran medida el resultado de la dilatación.  
- Para un elemento estructurante rectangular se observa que la dilatación es mayor en sentido vertical, dandole una apariencia de alargamiendo además del ensanchamiento.  
- Para un elemento estructurante eliptico se observa como la imagen además de su ensanchamiento por dilatación tiende a una forma rómbica.  
- Para un elemento estructurante en cruz se observa que además del ensanchamiento por dilatación también tienen a una forma más cuadratica con un giro en su centro de 45°, es decir, a una forma más en "cruz" con biseles claramente marcados para 250 iteraciones.

### 2. Erosión

```python
erosion_1 = cv2.erode(img_1_bin,kernel1,iterations=1)
erosion_2 = cv2.erode(img_2_bin,kernel1,iterations=1)
erosion_10 = cv2.erode(img_1_bin,kernel1,iterations=10)
erosion_20 = cv2.erode(img_2_bin,kernel1,iterations=10)
erosion_100 = cv2.erode(img_1_bin,kernel1,iterations=50)
erosion_200 = cv2.erode(img_2_bin,kernel1,iterations=100)

print('IMAGENES ORIGINALES')
plt.figure(1)
plt.subplot(1,2,1), plt.imshow(img_1_bin,cmap='gray'), plt.title('Mano 1')
plt.subplot(1,2,2), plt.imshow(img_2_bin,cmap='gray'), plt.title('Mano 2')
plt.subplots_adjust(right = 1.5)
plt.show()

print('SE EROSIONA CON ELEMENTO ESTRUCTURANTE CUADRADO: ')
plt.figure(2)
plt.subplot(1,2,1), plt.imshow(erosion_1,cmap='gray'), plt.title('#iteraciones = 1')
plt.subplot(1,2,2), plt.imshow(erosion_2,cmap='gray'), plt.title('#iteraciones = 1')
plt.subplots_adjust(right = 1.5)
plt.show()


plt.figure(3)
plt.subplot(1,2,1), plt.imshow(erosion_10,cmap='gray'), plt.title('#iteraciones = 10')
plt.subplot(1,2,2), plt.imshow(erosion_20,cmap='gray'), plt.title('#iteraciones = 10')
plt.subplots_adjust(right = 1.5)
plt.show()


plt.figure(4)
plt.subplot(1,2,1), plt.imshow(erosion_100,cmap='gray'), plt.title('#iteraciones = 50')
plt.subplot(1,2,2), plt.imshow(erosion_200,cmap='gray'), plt.title('#iteraciones = 100')
plt.subplots_adjust(right = 1.5)
plt.show()
```

Notas:  
- Como se observa ´erosionar´ con el elemento estructurante cudrado proboca un "desgaste" o explanción del fondo frente al contorno de interes.  
- Este desgaste depende en gran medida del número de iteraciones que se realize, como se observa para una sólo iteración la alteración entre la imagen original y la erosionada no es muy evidente, pero para 100 iteraciones el contorno esta totalmente perdido para la mano 1 y parcialmente perdido para la mano 2.  
- La erosión depende también en gran medida del elemento estructurante que se utilize, para ello se hará uso de los elementos estructurantes: ´kernel_rect´ ; ´kernel_elip´ y ´kernel_cruz´

```python
erosion_rect_1 = cv2.erode(img_2_bin,kernel_rect,iterations=1)
erosion_elip_1 = cv2.erode(img_2_bin,kernel_elip,iterations=1)
erosion_cross_1 = cv2.erode(img_2_bin,kernel_cross,iterations=1)
erosion_rect_2 = cv2.erode(img_2_bin,kernel_rect,iterations=10)
erosion_elip_2 = cv2.erode(img_2_bin,kernel_elip,iterations=10)
erosion_cross_2 = cv2.erode(img_2_bin,kernel_cross,iterations=10)
erosion_rect_3 = cv2.erode(img_2_bin,kernel_rect,iterations=25)
erosion_elip_3 = cv2.erode(img_2_bin,kernel_elip,iterations=25)
erosion_cross_3 = cv2.erode(img_2_bin,kernel_cross,iterations=25)
erosion_rect_4 = cv2.erode(img_2_bin,kernel_rect,iterations=50)
erosion_elip_4 = cv2.erode(img_2_bin,kernel_elip,iterations=50)
erosion_cross_4 = cv2.erode(img_2_bin,kernel_cross,iterations=50)


plt.figure(1)
plt.subplot(1,1,1), plt.imshow(img_2_bin,cmap='gray'),plt.title('imagen original')
plt.show()

plt.figure(2)
print('SE EROSIONA CON ELEMENTO ESTRUCTURANTE RECTANGULAR')
plt.subplot(1,4,1), plt.imshow(erosion_rect_1,cmap='gray'), plt.title('# Itereciones= 1');
plt.subplot(1,4,2), plt.imshow(erosion_rect_2,cmap='gray'), plt.title('# Itereciones= 10');
plt.subplot(1,4,3), plt.imshow(erosion_rect_3,cmap='gray'), plt.title('# Itereciones= 25');
plt.subplot(1,4,4), plt.imshow(erosion_rect_4,cmap='gray'), plt.title('# Itereciones= 50');
plt.subplots_adjust(right = 1.5)
plt.show()

plt.figure(3)
print('SE EROSIONA CON ELEMENTO ESTRUCTURANTE ELÍPTICO')
plt.subplot(1,4,1), plt.imshow(erosion_elip_1,cmap='gray'), plt.title('# Itereciones= 1');
plt.subplot(1,4,2), plt.imshow(erosion_elip_2,cmap='gray'), plt.title('# Itereciones= 10');
plt.subplot(1,4,3), plt.imshow(erosion_elip_3,cmap='gray'), plt.title('# Itereciones= 25');
plt.subplot(1,4,4), plt.imshow(erosion_elip_4,cmap='gray'), plt.title('# Itereciones= 50')
plt.subplots_adjust(right = 1.5)
plt.show()

plt.figure(4)
print('SE EROSIONA CON ELEMENTO ESTRUCTURANTE CRUZ')
plt.subplot(1,4,1), plt.imshow(erosion_cross_1,cmap='gray'), plt.title('# Itereciones= 1');
plt.subplot(1,4,2), plt.imshow(erosion_cross_2,cmap='gray'), plt.title('# Itereciones= 10');
plt.subplot(1,4,3), plt.imshow(erosion_cross_3,cmap='gray'), plt.title('# Itereciones= 25')
plt.subplot(1,4,4), plt.imshow(erosion_cross_4,cmap='gray'), plt.title('# Itereciones= 50');
plt.subplots_adjust(right = 1.5)
plt.show()
```

Notas:  
- Se observa que la erosión depende tanto de las iteraciones realizadas como del elemento estructurante. Siendo más evidente con el elemento estructurante rectangular que con los otros dos elementos estructurantes.
- Para un elemento estructurante rectangular el desgaste del fondo con respecto al contorno tiene una tendencia a ser mayor en el sentido vertical de la imagen. Como se observa en la mancha interna de la mano, la cual tiene a ser de manera rectangular para 50 iteraciones.  
- Para un elemento estructurante elíptico el desgaste es de manera más uniforme pero también es mayor, llegando a desaparecer por completo los dedos para 50 iteraciones y dejando una mancha casí uniforme en el interior de la mano.  
- Para un elemento estructurante en cruz se observa un desgaste similar que el obtenido con un elemento estructurante elíptico, sin embargo, con este elemento estructurante se obtiene un menor desgaste para 50 iteraciones lograndose observar algo de los dedos de la mano, también se observa que con este elemento estructurante se lográn biseles bien marcado en la imagen.
