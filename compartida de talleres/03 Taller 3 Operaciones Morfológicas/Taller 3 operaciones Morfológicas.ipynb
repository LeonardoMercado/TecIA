```python
import numpy as np
import cv2 
from matplotlib import pyplot as plt
print('Librerias importadas correctamente')
```

```python
img_1 = cv2.imread('Fig/manoP.png',0)
img_2 = cv2.imread('Fig/manoG.jpg',0)
ret,img_1_bin = cv2.threshold(img_1,127,255,cv2.THRESH_BINARY)
ret,img_2_bin = cv2.threshold(img_2,165,255,cv2.THRESH_BINARY_INV)
print('Las imagenes a trabajar son:')

plt.figure(1)
plt.subplot(2,2,1), plt.imshow(img_1,cmap='gray'),plt.title('Mano 1')
plt.subplot(2,2,2), plt.imshow(img_2,cmap='gray'),plt.title('Mano 2')
plt.subplot(2,2,3), plt.imshow(img_1_bin,cmap='gray'),plt.title('Mano 1 Binaria')
plt.subplot(2,2,4), plt.imshow(img_2_bin,cmap='gray'),plt.title('Mano 2 Binaria')
plt.subplots_adjust(left = 0.125, right = 1.5 ,  bottom = 0.1, top = 1.8, wspace = 0.2, hspace = 0.2 )
plt.show()
```

###### 1. Dilatación

```python
kernel1 = np.ones((3,3))
dilatation_1 = cv2.dilate(img_1_bin,kernel1,iterations=1)
dilatation_2 = cv2.dilate(img_2_bin,kernel1,iterations=1)
dilatation_10 = cv2.dilate(img_1_bin,kernel1,iterations=10)
dilatation_20 = cv2.dilate(img_2_bin,kernel1,iterations=10)
dilatation_100 = cv2.dilate(img_1_bin,kernel1,iterations=100)
dilatation_200 = cv2.dilate(img_2_bin,kernel1,iterations=100)

plt.figure(1)
plt.subplot(4,2,1), plt.imshow(img_1_bin,cmap='gray'), plt.title('Mano 1')
plt.subplot(4,2,2), plt.imshow(img_2_bin,cmap='gray'), plt.title('Mano 2')
plt.subplot(4,2,3), plt.imshow(dilatation_1,cmap='gray'), plt.title('Dilatacion 1 iteraciones')
plt.subplot(4,2,4), plt.imshow(dilatation_2,cmap='gray'), plt.title('Dilatacion 1 iteraciones')
plt.subplot(4,2,5), plt.imshow(dilatation_10,cmap='gray'), plt.title('Dilatacion 10 iteraciones')
plt.subplot(4,2,6), plt.imshow(dilatation_20,cmap='gray'), plt.title('Dilatacion 10 iteraciones')
plt.subplot(4,2,7), plt.imshow(dilatation_100,cmap='gray'), plt.title('Dilatacion 100 iteraciones')
plt.subplot(4,2,8), plt.imshow(dilatation_200,cmap='gray'), plt.title('Dilatacion 100 iteraciones')
plt.subplots_adjust(left = 0.125, right = 1 ,  bottom = 0.1, top = 1.8, wspace = 0.2, hspace = 0.4)
plt.show()
```
