### 1. Dilatación
La dilatación tiene como operador el siguiente símbolo:    

![nameimge](https://i.imgur.com/8YT3iAa.jpg)

```python
kernel1 = np.ones((3,3))
dilatation_1 = cv2.dilate(img_1_bin,kernel1,iterations=1)
dilatation_2 = cv2.dilate(img_2_bin,kernel1,iterations=1)
dilatation_10 = cv2.dilate(img_1_bin,kernel1,iterations=10)
dilatation_20 = cv2.dilate(img_2_bin,kernel1,iterations=10)
dilatation_100 = cv2.dilate(img_1_bin,kernel1,iterations=50)
dilatation_200 = cv2.dilate(img_2_bin,kernel1,iterations=100)

print('IMAGENES ORIGINALES')
plt.figure(1)
plt.subplot(1,2,1), plt.imshow(img_1_bin,cmap='gray'), plt.title('Mano 1')
plt.subplot(1,2,2), plt.imshow(img_2_bin,cmap='gray'), plt.title('Mano 2')
plt.subplots_adjust(right = 1.5)
plt.show()

print('SE DILATA CON ELEMENTO ESTRUCTURANTE CUADRADO: ')
plt.figure(2)
plt.subplot(1,2,1), plt.imshow(dilatation_1,cmap='gray'), plt.title('#iteraciones = 1')
plt.subplot(1,2,2), plt.imshow(dilatation_2,cmap='gray'), plt.title('#iteraciones = 1')
plt.subplots_adjust(right = 1.5)
plt.show()


plt.figure(3)
plt.subplot(1,2,1), plt.imshow(dilatation_10,cmap='gray'), plt.title('#iteraciones = 10')
plt.subplot(1,2,2), plt.imshow(dilatation_20,cmap='gray'), plt.title('#iteraciones = 10')
plt.subplots_adjust(right = 1.5)
plt.show()


plt.figure(4)
plt.subplot(1,2,1), plt.imshow(dilatation_100,cmap='gray'), plt.title('#iteraciones = 50')
plt.subplot(1,2,2), plt.imshow(dilatation_200,cmap='gray'), plt.title('#iteraciones = 100')
plt.subplots_adjust(right = 1.5)
plt.show()
```

Notas:  
- Como se observa el efecto de dilatar las imagenes con el elemento estructurante que denominamos para este caso kernel1, hace que la mano "se engrose", o mejor dicho, se extiendan las partes blancas sobre las negras.  
- Para la dilatación con una iteración se observa que para las dos manos se alcanza a cubrir parte de un "hueco" que tenia, y ya en 10 iteraciones estos "huecos" se cubren por completo. 
- Se observa también que a medida que se itera más, la dilatación hace que el contorno se ensanche más y más llegando a perderse por completo el contorno de los dedos. 

```python
kernel_rect = cv2.getStructuringElement(cv2.MORPH_RECT,(3,8),(1,1))
kernel_elip = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(8,3),(1,1))
kernel_cross = cv2.getStructuringElement(cv2.MORPH_CROSS,(5,5),(2,2))
print('Kerner rectangular')
print(kernel_rect)
print('----------')
print('Kerner elíptico')
print(kernel_elip)
print('----------')
print('Kerner Cruz')
print(kernel_cross)
```

En la siguiente etapa se examinará como afectan los distintos elementos estructurantes a la dilatación de las imagenes.  

```python
dilatation_rect_1 = cv2.dilate(img_2_bin,kernel_rect,iterations=1)
dilatation_elip_1 = cv2.dilate(img_2_bin,kernel_elip,iterations=1)
dilatation_cross_1 = cv2.dilate(img_2_bin,kernel_cross,iterations=1)
dilatation_rect_2 = cv2.dilate(img_2_bin,kernel_rect,iterations=10)
dilatation_elip_2 = cv2.dilate(img_2_bin,kernel_elip,iterations=10)
dilatation_cross_2 = cv2.dilate(img_2_bin,kernel_cross,iterations=10)
dilatation_rect_3 = cv2.dilate(img_2_bin,kernel_rect,iterations=100)
dilatation_elip_3 = cv2.dilate(img_2_bin,kernel_elip,iterations=100)
dilatation_cross_3 = cv2.dilate(img_2_bin,kernel_cross,iterations=100)
dilatation_rect_4 = cv2.dilate(img_2_bin,kernel_rect,iterations=250)
dilatation_elip_4 = cv2.dilate(img_2_bin,kernel_elip,iterations=250)
dilatation_cross_4 = cv2.dilate(img_2_bin,kernel_cross,iterations=250)


plt.figure(1)
plt.subplot(1,1,1), plt.imshow(img_2_bin,cmap='gray'),plt.title('imagen original')
plt.show()

plt.figure(2)
print('SE DILATA CON ELEMENTO ESTRUCTURANTE RECTANGULAR')
plt.subplot(1,4,1), plt.imshow(dilatation_rect_1,cmap='gray'), plt.title('# Itereciones= 1');
plt.subplot(1,4,2), plt.imshow(dilatation_rect_2,cmap='gray'), plt.title('# Itereciones= 10');
plt.subplot(1,4,3), plt.imshow(dilatation_rect_3,cmap='gray'), plt.title('# Itereciones= 100');
plt.subplot(1,4,4), plt.imshow(dilatation_rect_4,cmap='gray'), plt.title('# Itereciones= 250');
plt.subplots_adjust(right = 1.5)
plt.show()

plt.figure(3)
print('SE DILATA CON ELEMENTO ESTRUCTURANTE ELÍPTICO')
plt.subplot(1,4,1), plt.imshow(dilatation_elip_1,cmap='gray'), plt.title('# Itereciones= 1');
plt.subplot(1,4,2), plt.imshow(dilatation_elip_2,cmap='gray'), plt.title('# Itereciones= 10');
plt.subplot(1,4,3), plt.imshow(dilatation_elip_3,cmap='gray'), plt.title('# Itereciones= 100');
plt.subplot(1,4,4), plt.imshow(dilatation_elip_4,cmap='gray'), plt.title('# Itereciones= 250')
plt.subplots_adjust(right = 1.5)
plt.show()

plt.figure(4)
print('SE DILATA CON ELEMENTO ESTRUCTURANTE CRUZ')
plt.subplot(1,4,1), plt.imshow(dilatation_cross_1,cmap='gray'), plt.title('# Itereciones= 1');
plt.subplot(1,4,2), plt.imshow(dilatation_cross_2,cmap='gray'), plt.title('# Itereciones= 10');
plt.subplot(1,4,3), plt.imshow(dilatation_cross_3,cmap='gray'), plt.title('# Itereciones= 100')
plt.subplot(1,4,4), plt.imshow(dilatation_cross_4,cmap='gray'), plt.title('# Itereciones= 250');
plt.subplots_adjust(right = 1.5)
plt.show()
```

Nota:  
- Como se observa el elemento estructurante determina en gran medida el resultado de la dilatación.  
- Para un elemento estructurante rectangular se observa que la dilatación es mayor en sentido vertical, dandole una apariencia de alargamiendo además del ensanchamiento.

### 3. Apertura
La apertura es una erosión seguida por una dilatación y su operador se simboliza de la siguiente manera:

![img](https://i.imgur.com/iXDJe2l.png)  
La operación de apertura se puede considerar como un filtro para ruido tipo "sal", es decir, al erosionar primero lo que se hace es eliminar figuras que sean más pequeñas que el elemento estructurante y luego al dilatar intenta volver a la normalidad las figuras que no se eliminarón por la erosión, Eliminando así el ruido tipo sal. 

```python
import random
def sp_noise(image,prob):
       output = np.zeros(image.shape,np.uint16)
       thres = 1 - prob 
       for i in range(image.shape[0]):
           for j in range(image.shape[1]):
               rdn = random.random()
               if rdn < prob:
                   output[i][j] = 0
               elif rdn > thres:
                   output[i][j] = 255
               else:
                   output[i][j] = image[i][j]
       return output
```

Para observar el funcionamiento de la operación de apertura se le agregará ruido tipo sal y pimienta a la image de la mano con la función anterior

```python
img_noise = sp_noise(img_2_bin,0.1)

plt.figure(1)
plt.subplot(1,2,1), plt.imshow(img_2_bin,cmap='gray'),plt.title('imagen original')
plt.subplot(1,2,2), plt.imshow(img_noise,cmap='gray'),plt.title('imagen con ruido')
plt.subplots_adjust(right=1)
plt.show()
```

```python
kernel1 = np.ones((3,3))
apertura_1 = cv2.morphologyEx(img_noise,cv2.MORPH_OPEN,kernel1,iterations=1)
apertura_2 = cv2.morphologyEx(img_noise,cv2.MORPH_OPEN,kernel1,iterations=2)
apertura_3 = cv2.morphologyEx(img_noise,cv2.MORPH_OPEN,kernel1,iterations=3)

kernel2 = np.ones((5,5))
apertura_4 = cv2.morphologyEx(img_noise,cv2.MORPH_OPEN,kernel2,iterations=1)
apertura_5 = cv2.morphologyEx(img_noise,cv2.MORPH_OPEN,kernel2,iterations=2)
apertura_6 = cv2.morphologyEx(img_noise,cv2.MORPH_OPEN,kernel2,iterations=3)


print('IMAGENES ORIGINALES')
plt.figure(1)
plt.subplot(1,1,1), plt.imshow(img_noise,cmap='gray'), plt.title('Img Con ruido')
plt.subplots_adjust(right = 1.25)
plt.show()


print('APLICANDO LA APERTURA CON UN ELEMENTO ESTRUCTURANTE DE 3X3 CON UN NÚMERO DE ITERACIONES: ')
plt.figure(2)
plt.subplot(1,3,1), plt.imshow(apertura_1,cmap='gray'), plt.title('#iteraciones=1')
plt.subplot(1,3,2), plt.imshow(apertura_2,cmap='gray'), plt.title('#iteraciones=2')
plt.subplot(1,3,3), plt.imshow(apertura_3,cmap='gray'), plt.title('#iteraciones=3')
plt.subplots_adjust(right = 1.5)
plt.show()

print('APLICANDO LA APERTURA CON UN ELEMENTO ESTRUCTURANTE DE 5X5 CON UN NÚMERO DE ITERACIONES: ')
plt.figure(3)
plt.subplot(1,3,1), plt.imshow(apertura_4,cmap='gray'), plt.title('#iteraciones=1')
plt.subplot(1,3,2), plt.imshow(apertura_5,cmap='gray'), plt.title('#iteraciones=2')
plt.subplot(1,3,3), plt.imshow(apertura_6,cmap='gray'), plt.title('#iteraciones=3')
plt.subplots_adjust(right = 1.5)
plt.show()
```

Notas:  
- Como se puede observar realizar una erosión seguida de una dilatación hace la función de filtro de sal en la imagen de la mano.Quedando únicamente el ruido tipo pimienta.  
- También se observa como la apertura depende en gran medida del número de iteraciones así como del elemento estructurante.  
- Para una apertura con un elemento estructurante de 3x3 lleno de unos se observa que para una iteración el ruido de sal desaparece completamente dejando únicamente el ruido tipo pimienta al interior de la mano. También es notorio que a medida que aumenta el número de iteraciones el filtro es más fuerte con el contorno de la mano llagando a eliminarlo casi por completo con tan sólo 3 iteraciones.  
- Para una apertura con un elemento estructurante de 5x5 lleno de unos se observa que para una iteración se obtiene el mismo resultado que con un elemento estructurante 3x3 lleno de uno con dos iteraciones. Y ya con dos iteraciones la imagen de la mano a desaparecido completamente.

### 4. Cierre
El cierre es una dilatación seguida por una erosión y su operador se simboliza de la siguiente manera:  
![img](https://i.imgur.com/ptJyeIN.png)  
La operación de cierre se puede considerar como un filtro para ruido tipo "pimienta", es decir, al realizar primero la operación de dilatación provoca que los puntos negros sobre fondo blanco se ensanchen o rellenen, pero también hará que el ruido de puntos blancos sobre fondo negro aumente, para ello se realiza la operación de erosión el cual trata de eliminar o o disminuir el ensanchamiento de estos puntos blancos.


```python
kernel1 = np.ones((3,3))
cierre_1 = cv2.morphologyEx(img_noise,cv2.MORPH_CLOSE,kernel1,iterations=1)
cierre_2 = cv2.morphologyEx(img_noise,cv2.MORPH_CLOSE,kernel1,iterations=2)
cierre_3 = cv2.morphologyEx(img_noise,cv2.MORPH_CLOSE,kernel1,iterations=3)

kernel2 = np.ones((5,5))
cierre_4 = cv2.morphologyEx(img_noise,cv2.MORPH_CLOSE,kernel2,iterations=1)
cierre_5 = cv2.morphologyEx(img_noise,cv2.MORPH_CLOSE,kernel2,iterations=2)
cierre_6 = cv2.morphologyEx(img_noise,cv2.MORPH_CLOSE,kernel2,iterations=3)


print('IMAGENES ORIGINALES')
plt.figure(1)
plt.subplot(1,1,1), plt.imshow(img_noise,cmap='gray'), plt.title('Img Con ruido')
plt.subplots_adjust(right = 1.25)
plt.show()


print('APLICANDO EL CIERRE CON UN ELEMENTO ESTRUCTURANTE DE 3X3 CON UN NÚMERO DE ITERACIONES: ')
plt.figure(2)
plt.subplot(1,3,1), plt.imshow(cierre_1,cmap='gray'), plt.title('#iteraciones=1')
plt.subplot(1,3,2), plt.imshow(cierre_2,cmap='gray'), plt.title('#iteraciones=2')
plt.subplot(1,3,3), plt.imshow(cierre_3,cmap='gray'), plt.title('#iteraciones=3')
plt.subplots_adjust(right = 1.5)
plt.show()

print('APLICANDO EL CIERRE CON UN ELEMENTO ESTRUCTURANTE DE 5X5 CON UN NÚMERO DE ITERACIONES: ')
plt.figure(3)
plt.subplot(1,3,1), plt.imshow(cierre_4,cmap='gray'), plt.title('#iteraciones=1')
plt.subplot(1,3,2), plt.imshow(cierre_5,cmap='gray'), plt.title('#iteraciones=2')
plt.subplot(1,3,3), plt.imshow(cierre_6,cmap='gray'), plt.title('#iteraciones=3')
plt.subplots_adjust(right = 1.5)
plt.show()
```

Notas:  
- Como se observa aplicar la operación de cierre provoco que el ruido tipo pimineta que estaba presente al interior de la mano desapareciera de manera considerable.  
- También se observa que esta operación aumentó el ruido tipo salta externo al contorno de la mano.  
- De igualmanera que con las otras operaciones la operación de cierre depende en gran medida del elemento estructurante utilizado así como de las iteraciones realizadas.  
- Para un elemento estructurante de 3x3 lleno de unos y con una sola iteración se logra eliminar casi por completo el ruido tipo pimienta en el interior de la mano, y para dos iteraciones se ha eliminado el ruido por completo. Sin embargo, también implica que el ruido tipo sal al exterior del contorno de la mano aumente de manera apreciable. y para 3 iteraciones el contorno de la mano es apenas apreciable.  
- Para un elemento estructurante de 5x5 lleno de unos con una sola iteración se logra lo mismo que con un elemento estructurante 3x3 lleno de unos con 2 iteraciones. Sin embargo, para 2 iteraciones ya el contorno de la mano se ha perdido por completo.

### 5. Otras operaciones

La biblioteca de `Open CV` nos permite realizar otras operaciones tales como:    

1. Gradiente.   
2. TopHat.  
3. BlackHat.

```python
kernel1 = np.ones((3,3))
gradient = cv2.morphologyEx(img_2_bin,cv2.MORPH_GRADIENT,kernel1,iterations=10)
top_hat = cv2.morphologyEx(img_2_bin,cv2.MORPH_TOPHAT,kernel1,iterations=65)
black_hat = cv2.morphologyEx(img_2_bin,cv2.MORPH_BLACKHAT,kernel1,iterations=25)



plt.figure(1)
plt.subplot(1,2,1), plt.imshow(img_2_bin,cmap='gray'),plt.title('imagen original')
plt.subplots_adjust(right=1)
plt.show()

print('APLICANDO LA OPERACIÓN DE')
plt.figure(2)
plt.subplot(1,3,1), plt.imshow(gradient,cmap='gray'), plt.title('Gradiente')
plt.subplot(1,3,2), plt.imshow(top_hat,cmap='gray'), plt.title('Top Hat')
plt.subplot(1,3,3), plt.imshow(black_hat,cmap='gray'), plt.title('Black Hat')
plt.subplots_adjust(right = 1.5)
plt.show()
```

Notas:  
- Como se observa aplicar una operación de gradiente retorna el borde la de mano.  
- Como se observa aplicar una operación de Tophat retorna el borde superior y lateral izquierdo de la imagen.  
- Como se observa aplicar una operación de Blackhay retorna el borde inferior y lateral derecho de la imagen.

### 6. Filtro Morfológico

Como se ha mencionado anteriormente aplicar una operación de apertura hace la función de un filtro de sal, y aplicar una operación de cierre hace la función de un filtro de pimienta, aplicando estas dos operaciones vamos a obtener un filtro de sal y pimienta. Es decir:  
![img](https://i.imgur.com/WIwIKwJ.png)

```python
kernel1 = np.ones((3,3))
apertura_1 = cv2.morphologyEx(img_noise,cv2.MORPH_OPEN,kernel1)
apertura_1 = cv2.morphologyEx(apertura_1,cv2.MORPH_CLOSE,kernel1,iterations=3)

plt.figure(1)
plt.subplot(1,3,1), plt.imshow(img_2_bin,cmap='gray'), plt.title('Imagen Original')
plt.subplot(1,3,2), plt.imshow(img_noise,cmap='gray'), plt.title('Imagen con ruido')
plt.subplot(1,3,3), plt.imshow(apertura_1,cmap='gray'), plt.title('Imagen filtrada por Apertura y cierre')
plt.subplots_adjust(right = 2)
plt.show()
```

Notas:  
- Como se observa aplicar una apertura seguida de un cierre tiene la función de filtro de ruido tipo sal y pimienta.
