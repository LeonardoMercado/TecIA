```python
print('Por favor escoger la cantidad de BIN\'S')

@interact
def bins_valor(bin=[256,128,64,32,16,8,4,2]):
    global hist_light
    global hist_medium
    global hist_low
    
    hist_light = cv2.calcHist([image_light],[0],None,[bin],[0,256])
    hist_medium = cv2.calcHist([image_medium],[0],None,[bin],[0,256])
    hist_low = cv2.calcHist([image_low],[0],None,[bin],[0,256])

    correlation_light_light = cv2.compareHist(hist_light,hist_light, method=cv2.HISTCMP_CORREL)
    correlation_light_medium = cv2.compareHist(hist_light,hist_medium, method=cv2.HISTCMP_CORREL)
    correlation_light_low = cv2.compareHist(hist_light,hist_low, method=cv2.HISTCMP_CORREL)
    correlation_medium_medium = cv2.compareHist(hist_medium,hist_medium, method=cv2.HISTCMP_CORREL)
    correlation_medium_low = cv2.compareHist(hist_medium,hist_low, method=cv2.HISTCMP_CORREL)
    correlation_low_low = cv2.compareHist(hist_low,hist_low, method=cv2.HISTCMP_CORREL)

    print('Comparando los histogramas sin normalizar por el método de la correlación se obtiene: ')
    print('ligh con light %2.3f'%(correlation_light_light))
    print('light con medium %2.3f'%(correlation_light_medium))
    print('ligh con low %2.3f'%(correlation_light_low))
    print('medium con medium %2.3f'%(correlation_medium_medium))
    print('medium con low %2.3f'%(correlation_medium_low))
    print('low con low %2.3f'%(correlation_low_low))
```

#### Normalizando los histogramas: 


Para la normalización de los histogramas se ha recurrido a diversos métodos de la biblioteca de open Cv, desafortunadamente, estos metodos no dan resultados coherentes al momento de realizar las comparaciones, por ello se recure a la siguiente función:

```python
def normalize_histogram(histogram):
    v_max = histogram.max()
    v_min = histogram.min()
    return (histogram - v_min)/(v_max - v_min)
```

```python
@interact
def ploteo_hist(bin=[256,128,64,32,16,8,4,2]):
    hist_light = cv2.calcHist([image_light],[0],None,[bin],[0,256])
    hist_medium = cv2.calcHist([image_medium],[0],None,[bin],[0,256])
    hist_low = cv2.calcHist([image_low],[0],None,[bin],[0,256])   
        
    hist_light_normalize = normalize_histogram(hist_light)
    hist_medium_normalize = normalize_histogram(hist_medium)
    hist_low_normalize = normalize_histogram(hist_low)
    
    plt.figure(1)
    plt.subplot(131), plt.plot(range(bin),hist_light_normalize), plt.title('Hist Ligh')
    plt.subplot(132), plt.plot(range(bin),hist_medium_normalize), plt.title('Hist Ligh')
    plt.subplot(133), plt.plot(range(bin),hist_low_normalize), plt.title('Hist Ligh')

    plt.subplots_adjust(top=0.9, bottom=0.1, left=0.10, right=0.95, hspace=0.25, wspace=0.5)

    plt.show()
```

Se trabaja con tres imagenes del mismo elemento con condiciones de iluminación mas marcadas (interior, exterior soleado, exterior con sombra),
y el tamaño se mantiene igual en los tres casos procurando que los resultados obtenidos sean basados en datos uniformes.

```python
# SE CARGAN LAS IMAGENES 
imgInt = cv2.imread('Fig/Interior.jpeg',0)  # Carga imagen en grises
imgSol = cv2.imread('Fig/Sol.jpeg',0)       # Carga imagen en grises
imgSombra = cv2.imread('Fig/Sombra.jpeg',0) # Carga imagen en grises
plt.figure(1)
plt.subplot(131), plt.imshow(imgInt, cmap='gray'), plt.title('Interior');
plt.subplot(132), plt.imshow(imgSol, cmap='gray'), plt.title('Sol');
plt.subplot(133), plt.imshow(imgSombra, cmap='gray'), plt.title('Sombra');
plt.subplots_adjust(right=1.5)
```

```python
#CALCULO DE HISTOGRAMAS NO NORMALIZADOS
bins=[256,16,2]
Comp=np.zeros((3,3, 4)) 
CompNorm=np.zeros((3,3, 4)) 
for i in range(len(bins)):
    histInt = cv2.calcHist([imgInt],[0],None,[bins[i]],[0,256])    
    histSol = cv2.calcHist([imgSol],[0],None,[bins[i]],[0,256])
    histSombra = cv2.calcHist([imgSombra],[0],None,[bins[i]],[0,256])    

    Metodos=['Correlación', 'Chicuadrado', 'Intersección', 'Distancia Bhattacharyya']    
    
    #COMPARACIÓN NO NORMALIZADOS
    for metodo in range(len(Metodos)):
        Comp[i,0,metodo]=cv2.compareHist(histInt,histSol,metodo) 
    for metodo in range(4):
        Comp[i,1,metodo]=cv2.compareHist(histInt,histSombra,metodo)
    for metodo in range(4):
        Comp[i,2,metodo]=cv2.compareHist(histSol,histSombra,metodo)

    #PLOT HISTOGRAMAS NO NORMALIZADOS
    plt.figure(bins[i])
    fig, ax = plt.subplots();
    fig.suptitle(['Histograma no Normalizado', bins[i], 'bins'] , fontsize=16, x=1);
    plt.subplot(131),plt.plot(histInt), plt.title('Interior')
    plt.subplot(132),plt.plot(histSol), plt.title('Sol')
    plt.subplot(133),plt.plot(histSombra), plt.title('Sombra')
    plt.subplots_adjust(top=0.85,right=2)
    plt.show()
    
    histIntNorm=histInt
    histSolNorm=histSol
    histSombraNorm=histSombra
    cv2.normalize( histInt, histIntNorm, 0, 1, cv2.NORM_MINMAX);
    cv2.normalize( histSol, histSolNorm, 1e-12, 1, cv2.NORM_MINMAX);
    cv2.normalize( histSombra, histSombraNorm, 0, 1, cv2.NORM_MINMAX);
    
    #COMPARACIÓN NORMALIZADOS
    for metodo in range(len(Metodos)):
        CompNorm[i,0,metodo]=cv2.compareHist(histIntNorm,histSolNorm,metodo) 
    for metodo in range(4):
        CompNorm[i,1,metodo]=cv2.compareHist(histIntNorm,histSombraNorm,metodo) 
    for metodo in range(4):
        CompNorm[i,2,metodo]=cv2.compareHist(histSolNorm,histSombraNorm,metodo)

    #PLOT HISTOGRAMAS NORMALIZADOS
    plt.figure(bins[i])
    fig, ax = plt.subplots();
    fig.suptitle(['Histograma Normalizado', bins[i], 'bins'] , fontsize=16, x=1);
    plt.subplot(131),plt.plot(histIntNorm), plt.title('Interior')
    plt.subplot(132),plt.plot(histSolNorm), plt.title('Sol')
    plt.subplot(133),plt.plot(histSombraNorm), plt.title('Sombra')
    plt.subplots_adjust(top=0.85,right=2)
    plt.show()
```

Podemos concluir que con dos bins no es posible obtener información relevante respecto a la distribución de
grises en la imagen salvo, tal vez, una indicación de hacia que extremo se concentran la mayor parte de los datos.
Como es de esperarse, un aumento en el número de bins permite obtener información mas detallada respecto a la
distribución de grises en la imagen. Así, con 16 bins el histograma muestra los picos mas relevantes de concentración
de grises y con 255 bins es posible identificar cada componenete de grises en detalle.

## Análisis y Conclusiones
- Se observa que, para los histogramas no normalizados en las comparaciones por Chi-Cuadrado e Intersección, se obtienen valores altos que no dan claridad sobre las variaciones en esta relación respecto al número de bins o al tipo de iluminación de la imagen. Caso contrario ocurre con las comparaciones por correlación o distancia Bhattacharyya. 
- La correlación, al mostrar la relación lineal entre las dos variables no varía con la normalización, en la medida en que la normalización se podría considerar como una escalización de la gráfica de cada histograma, sin modificar esto la relación lineal entre los datos comparados.  
- En la comparación por distancia Bhattacharyya se observa el mismo comportamiento para histogramas normalizados y no normalizados debido a que el cálculo del índice lleva a la normalización de los datos.
- Las discrepancias observadas en los histogramas normalizados de dos bins, que  muestran una relaciones de 0, 1 o -1 se debe a que la información contenida en el diagrama normalizado destaca únicamente la tendencia general de los datos, así, al comparar dos imágenes con mayor tendencia hacia los tonos claros (sol vs interior) con la distancia Bhattacharyya. se obtine un cero (lo que correspondería a una pareja perfecta) y al comparar una imagen iluminada con una oscura por el mismo método (sol vs sombra)  se obtiene un 1 lo que se corresponde con el hecho de que tienen tendencias opuestas.
- En los histogramas de dos bins se tiene que para imágenes con la mayor parte de los pixeles hacia el mismo extremo del espectro se obtendrá una correlación de 1, una relación Chi-Cuadrado de 0, una interseción de 1 y una distancia Dist_Bhattacharyya de cero; mientras que, para datos con tendencias opuestas se tendra  -1 en correlación, 1 en Chi-Cuadrado, 0 en Intersección y 1 en distancia Bhattacharyya. Valores que en lugar de interpretarse como datos coincidentes o comportamentos diametralmente opuestos se ven como tendencias generales.
- Como se mencionó anteriormente el aumento en el número de bins mejora el cálculo de la correlación, sin embargo, los datos obtenidos para 16 bins ya dan un buen acercamiento a la similitud entre las gráficas.
- Como es de esperarse la correlación en todos los casos tiende a cero, lo que muestra que no hay una relación lineal entre los histogramas de las tres imágenes, salvo en los casos mencionados para dos bins.
- Incluso con los datos normalizados, la comparación por Chi-Cuadrado muestra altos desajustes entre los histográmas de las imágenes, lo que es de esperar teniendo en cuenta que las condiciones de iluminación no son similares.
