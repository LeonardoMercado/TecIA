<h1>Función: <span style="color:blue">cv2.filter2D</span></h1>

#http://docs.opencv.org/2.4/modules/imgproc/doc/filtering.html

#http://docs.opencv.org/2.4/modules/imgproc/doc/filtering.html#filter2d

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

from scipy.signal import convolve2d as conv2

from skimage import data, io, filters
```

```python
img = cv2.imread('fig/Notebook/Sudoku.jpg',0)

#Promedio
kernel1 = np.ones((3, 3)) / 9
kernel2 = np.ones((5, 5)) / 25
kernel3 = np.ones((7, 7)) / 49
kernel4 = np.ones((9, 9)) / 81
#kernel5 = np.ones((15, 15)) / 225

#Convolucion
imgC1 = conv2(img, kernel1, 'same')
imgC2 = conv2(img, kernel2, 'same')
imgF2 = cv2.filter2D(img,-1,kernel2)
imgF3 = cv2.filter2D(img,-1,kernel3)
imgF4 = cv2.filter2D(img,-1,kernel4)

#imgC3 = conv2(img, kernel3, 'same')
#imgC4 = conv2(img, kernel4, 'same')
#imgC5 = conv2(img, kernel5, 'same')




#VISUALIZACION
plt.figure(1)
plt.subplot(1,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2),plt.imshow(imgC1,cmap = 'gray')
plt.title('Conv 3X3'), plt.xticks([]), plt.yticks([])

plt.figure(2)
plt.subplot(1,2,1),plt.imshow(imgC2,cmap = 'gray')
plt.title('Conv 5X5'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2),plt.imshow(imgF2,cmap = 'gray')
plt.title('Filter 5X5'), plt.xticks([]), plt.yticks([])

plt.figure(3)
plt.subplot(1,2,1),plt.imshow(imgF3,cmap = 'gray')
plt.title('Filter 7X7'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2),plt.imshow(imgF4,cmap = 'gray')
plt.title('Filter 9X9'), plt.xticks([]), plt.yticks([])

plt.show()
```

<h1>Función: <span style="color:blue">cv2.medianBlur</span></h1>

```python
img = cv2.imread('fig/Notebook/Sudoku.jpg',0)

#Promedio
kernel2 = np.ones((5, 5)) / 25
imgC2 = conv2(img, kernel2, 'same')

#Median
imgM = cv2.medianBlur(img,5)


#VISUALIZACION
plt.figure(1)
plt.subplot(1,2,1),plt.imshow(imgC2,cmap = 'gray')
plt.title('Promedio 5X5'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2),plt.imshow(imgM,cmap = 'gray')
plt.title('Mediana 5X5'), plt.xticks([]), plt.yticks([])


plt.show()
```

<h1>Función: <span style="color:blue">cv2.Sobel</span></h1>    

```python
img = cv2.imread('fig/Notebook/Sudoku.jpg',0)

laplacian = abs(cv2.Laplacian(img,cv2.CV_64F))
sobelx = abs(cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5))
sobely = abs(cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5))

#VISUALIZACION
plt.figure(1)
plt.subplot(1,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])

plt.figure(2)
plt.subplot(1,2,1),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img = cv2.imread('fig/Notebook/Sudoku.jpg',0)

# Output dtype = cv2.CV_8U
sobelx8u = cv2.Sobel(img,cv2.CV_8U,1,0,ksize=5)

# Output dtype = cv2.CV_64F. Then take its absolute and convert to cv2.CV_8U
sobelx64f = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
abs_sobel64f = np.absolute(sobelx64f)
sobel_8u = np.uint8(abs_sobel64f)


plt.subplot(1,2,1),plt.imshow(sobelx8u,cmap = 'gray')
plt.title('Sobel CV_8U'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2),plt.imshow(sobel_8u,cmap = 'gray')
plt.title('Sobel abs(CV_64F)'), plt.xticks([]), plt.yticks([])

plt.show()

```

<h1>Función: <span style="color:blue">cv2.Canny</span></h1>  

```python
#Canny
#https://en.m.wikipedia.org/wiki/Canny_edge_detector


img = cv2.imread('fig/Notebook/Sudoku.jpg',0)
edges = cv2.Canny(img,100,200)


plt.figure(1)
plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
#Convolve

#https://docs.scipy.org/doc/scipy-0.18.1/reference/generated/scipy.signal.convolve2d.html

#scipy.signal.convolve2d(in1, in2, mode='full', boundary='fill', fillvalue=0)
```

```python
from scipy import signal
from scipy import misc



#ascent = misc.ascent()
ascent = cv2.imread('fig/Notebook/Sudoku.jpg',0)
scharr = np.array([[ -3-3j, 0-10j,  +3 -3j],
                    [-10+0j, 0+ 0j, +10 +0j],
                   [ -3+3j, 0+10j,  +3 +3j]]) # Gx + j*Gy
grad = signal.convolve2d(ascent, scharr, boundary='symm', mode='same')
```

```python
import matplotlib.pyplot as plt
fig, (ax_orig, ax_mag, ax_ang) = plt.subplots(3, 1, figsize=(6, 15))
ax_orig.imshow(ascent, cmap='gray')
ax_orig.set_title('Original')
ax_orig.set_axis_off()
ax_mag.imshow(np.absolute(grad), cmap='gray')
ax_mag.set_title('Gradient magnitude')
ax_mag.set_axis_off()
ax_ang.imshow(np.angle(grad), cmap='hsv') # hsv is cyclic, like angles
ax_ang.set_title('Gradient orientation')
ax_ang.set_axis_off()
fig.show()
```

```python
import numpy as np
from scipy.misc import imread
from skimage import color, data, restoration
from scipy.signal import convolve2d as conv2


img = cv2.imread('fig/Notebook/Sudoku.jpg',0)
image = img
psf = np.ones((5, 5)) / 25
image = conv2(image, psf, 'same')
image += 0.1 * image.std() * np.random.standard_normal(image.shape)

deconvolved = restoration.wiener(image, psf, 0.00001)
#print deconvolved
#print image


plt.imshow(img,cmap = 'gray')
plt.title('img'), plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(image,cmap = 'gray')
plt.title('image Noise'), plt.xticks([]), plt.yticks([])
plt.show()
plt.subplot(1,2,2),plt.imshow(deconvolved,cmap = 'gray')
plt.title('deconvolved'), plt.xticks([]), plt.yticks([])

plt.show()
```
